```python
import numpy as np
import pandas as pd

import bokeh_catplot

import bokeh.io
import bokeh.plotting

bokeh.io.output_notebook()
```

```python
df = pd.read_csv('C:/Users/yang1/git/bootcamp/data/ctx_lineplot_1july2020.csv',na_values='*')
with open('C:/Users/yang1/git/bootcamp/data/ctx_lineplot_1july2020.csv') as f:
    for _ in range(100): #first 5 lines
        print(f.readline())
        
        df1 = df.replace(np.nan, '', regex=True)
```

```python
s1 = df1.loc[df['Subject']==1, ['ID', 'ctx (pg/ml)']]
s2 = df1.loc[df['Subject']==2, ['ID', 'ctx (pg/ml)']]
s3 = df1.loc[df['Subject']==3, ['ID', 'ctx (pg/ml)']]

```

```python
# create the figure , stored in variable "p"
p = bokeh.plotting.figure(
    frame_width=400,
    frame_height=300,
    x_axis_label='time',
    y_axis_label='ctx (pg/ml)',
   
    
)
    
p.line (
    source = s1,
    x='ID',
    y='ctx (pg/ml)',
    line_width =2,
    color ='grey',
    legend_label='1',
)
p.line (

    source = s2,
    x='ID',
    y='ctx (pg/ml)',
    line_width =2,
    color ='navy',
    legend_label='2',
    
)
p.line (
    source = s3,
    x='ID',
    y='ctx (pg/ml)',
    line_width =2,
    color ='orange',
    legend_label='3',
)


p.legend.location = 'top_right'
p.legend.click_policy = 'hide'

bokeh.io.show(p)
```
